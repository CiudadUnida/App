```
!pip install numpy matplotlib

import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la simulación
n_steps = 300  # Número de pasos
step_length = 1.0  # Longitud del paso
beta = 0.9  # Velocidad relativa de la partícula

# Generar trayectoria
theta = np.random.uniform(0, np.pi, n_steps)
phi = np.random.uniform(0, 2 * np.pi, n_steps)

x = step_length * np.sin(theta) * np.cos(phi)
y = step_length * np.sin(theta) * np.sin(phi)
z = np.cumsum(step_length * np.cos(theta))

# Calcular ángulo Cherenkov
sin_theta_c = 1 / (beta * 1.333)  # Índice de refracción del medio (ejemplo: agua)
theta_c = np.arcsin(sin_theta_c)

# Filtrar trayectoria en el ángulo Cherenkov
mask = theta < theta_c
x_cherenkov = x[mask]
y_cherenkov = y[mask]
z_cherenkov = z[mask]

# Visualizar trayectoria y ángulo Cherenkov
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot(x, y, z, label='Trayectoria')
ax.scatter(x_cherenkov, y_cherenkov, z_cherenkov, c='r', label='Radiación Cherenkov')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()

plt.show()
```
